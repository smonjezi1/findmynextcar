```python
def df_split3(string):
    try:
        string=string.title()
        string=re.sub(' +', ' ',string)
        string=string.strip(' ')
        regex1=re.compile(r'mercedes benz|mercedes[^benz]|mercedes$',flags=re.IGNORECASE)
        regex2=re.compile(r'land rover',flags=re.IGNORECASE)
        regex3=re.compile(r'alpha romeo',flags=re.IGNORECASE)
        regex4=re.compile(r'aston martin',flags=re.IGNORECASE)
        regex5=re.compile(r'rolls royce',flags=re.IGNORECASE)
        regex6=re.compile(r'([0-9])(\s+series)',flags=re.IGNORECASE)
        regex7=re.compile(r'[\*\!\'\"\$\@\%\^\&\~\`\-\=\_\+\\\|\/\?\(\)\{\}\[\]\;\:\,]')
        string=re.sub(regex7, ' ',string)
        string=re.sub(' +', ' ',string)
        string=string.strip(' ')
        string=re.sub(regex1, 'MercedesBenz',string)
        string=re.sub(regex2, 'LandRover',string)
        string=re.sub(regex3, 'AlphaRomeo',string)
        string=re.sub(regex4, 'AstonMartin',string)
        string=re.sub(regex5, 'RollsRoyce',string)
        string=regex6.sub(r'\1Series',string)
        string=re.sub(' +', ' ',string)
        string=string.strip(' .')
        strs=string.split(' ')
    except:
        return None
    try:
        int(strs[1])
        try:
            return ' '.join(strs[4:]) if ' '.join(strs[4:])!='' else None
        except:
            return None
    except:
        try:
            return ' '.join(strs[3:]) if ' '.join(strs[3:])!='' else None
        except:
            return None
        
    try:
        return ' '.join(strs[3:]) if ' '.join(strs[3:])!='' else None
    except:
        return None
df['trim0'][0:len(df['model'])]=list(map(df_split3, df['model'][0:len(df['model'])]))
```
