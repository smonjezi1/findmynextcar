```python
from fuzzywuzzy import fuzz
import re
def fill_all(string,string2):
    try:
        string=re.sub(r'4runner','4-Runner',string2,flags=re.IGNORECASE)
        string2=re.sub(r'4runner','4-Runner',string2,flags=re.IGNORECASE)
        string=string.title()
        string2=string2.title()
        regex1=re.compile(r'\bmercedes.*?benz\b|mercedes[^benz]|mercedes$',flags=re.IGNORECASE)
        if re.search(regex1, string) or re.search(regex1, string2):
            regex2=re.compile(r'\b([a-zA-Z]{1,3})(\d{2,3})\b',flags=re.IGNORECASE)
            string=regex2.sub(r'\1 \2',string)
            string2=regex2.sub(r'\1 \2',string2)
            string=re.sub(regex1, ' Mercedes-Benz ',string)
            string2=re.sub(regex1, ' Mercedes-Benz ',string2)

        regex3=re.compile(r'\bFord\b',flags=re.IGNORECASE)
        if re.search(regex3, string) or re.search(regex3, string2):
            regex4=re.compile(r'\b([a-zA-Z]{1})(\d{3})\b',flags=re.IGNORECASE)
            string=regex4.sub(r'\1-\2',string)
            string2=regex4.sub(r'\1-\2',string2)

        string=re.sub(' +', ' ',string)        
        string=string.strip(' ')
        string2=re.sub(' +', ' ',string2)
        string2=string2.strip(' ')
        year_return=None        
        make_return=None
        model_return=None
        for word in re.split('\s|,|\-',string):
            try:
                if (int(word) <2019) & (int(word) >= 1900):
                    year_return=int(word)
                    break
            except:
                continue
        if year_return==0:
            for word in re.split('\s|,|\-',string2):
                try:
                    if (int(word) <2019) & (int(word) >= 1900):
                        year_return=int(word)
                        break
                except:
                    continue
        mk=0
        mdl=0
        makes=list(d_make_models_trimed.keys())
        n_makes=len(makes)
        for i in range(n_makes):
            make=makes[i]
            if fuzz.token_set_ratio(make,string)>90:
                make_return=make
                mk=1
                models=d_make_models_trimed[make]
                n_models=len(d_make_models_trimed[make])
                for j in range(n_models):
                    if fuzz.token_set_ratio(models[j],string)>90:
                        model_return=models[j]
                        mdl=1
                        break
                if mdl==0:
                    for j in range(n_models):
                        if fuzz.token_set_ratio(models[j],string2)>90:
                            model_return=models[j]
                            mdl=1
                            break
                break
            if mk==0:
                if fuzz.token_set_ratio(make,string2)>90:
                    make_return=make
                    mk=1
                    models=d_make_models_trimed[make]
                    n_models=len(d_make_models_trimed[make])
                    for j in range(n_models):
                        if fuzz.token_set_ratio(models[j],string2)>90:
                            model_return=models[j]
                            mdl=1
                            break
                        if mdl==0:
                            for j in range(n_models):
                                if fuzz.token_set_ratio(models[j],string)>90:
                                    model_return=models[j]
                                    mdl=1
                                    break
                    break
        return year_return,make_return,model_return
    except:
        return None,None,None
```
